```python
# Libraries 

import matplotlib.image as mpimg
import matplotlib.pyplot as plt 
```

```python
# load image 
# Image has been converted to a numpy array

img = mpimg.imread('puppy.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
img
```

```python
# Display array back to image 

plt.imshow(img)
plt.show()
```

```python
# Resizing image - Using pillow 

from PIL import Image 

img = Image.open('puppy.jpg')
img_resized = img.resize((200, 200))
```

```python
plt.imshow(img_resized)
plt.show()
```

```python
img_resized.save('dog_image_resized.jpg')
```

```python
# Converting OpenCV library 
import cv2 

img = cv2.imread("puppy.jpg")
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
grayscale_image.shape
```

```python
cv2.imshow('image', grayscale_image)
```

```python
# saving the grayscale image 

cv2.imwrite('puppy_grayscale.jpg', grayscale_image)
```
