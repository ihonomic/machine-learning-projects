```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn import svm
from sklearn.metrics import accuracy_score 
```

```python
# status - 1 - Parkinson's, 0 - healthy
```

```python
df = pd.read_csv("parkinsons.csv")
```

```python
df
```

```python
df.drop(columns="name", axis=1, inplace=True)
```

```python
df.head()
```

```python
# Count how many people have the disease or not 
df['status'].value_counts()
```

```python
# Grouping the data based on the target/label
```

```python
df.groupby('status').mean()
```

```python
X = df.drop(columns="status", axis=1)
y = df["status"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.head()
```

Data standardization

This is the act of randing up the values to a range of values. usually btw 0 & 1. To 
remove outliners differences 

scaler = StandardScaler()

```python
scaler.fit(X_train)
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

Model Training 

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation 

```python
# Accuaracy score on training data 

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)

training_data_accuracy
```

```python
# Accuaracy score on test data 

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)

test_data_accuracy
```
